```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x), color='red', linestyle='-.')

#set limits for axes
plt.xlim(-1,10)
plt.ylim(-2, 2)

plt.axis('equal')
```

```python
plt.plot(x, np.sin(x), '-b', label='Sine wave')
plt.plot(x, np.cos(x), '--g', label='Cosinus wave')
plt.axis('equal')

plt.legend()
```

```python
#object oriented interface to plotting
ax = plt.axes()
ax.plot(x, np.cos(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
              xlabel='x', ylabel='y',
              title='cos(x)')
```
