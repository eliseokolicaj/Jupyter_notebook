```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filename = 'Data/California_cities.csv'
cities = pd.read_csv(filename)
cities.head()
```

```python
#Extract the information we need
lon, lat = cities['longd'], cities['latd']
population, area = cities['population_total'], cities['area_total_km2']

# Scatter the points, using size and color but no label
plt.scatter(lon, lat, label=None, c=np.log10(population), s=area, cmap='viridis', alpha=0.5, linewidth=0)
plt.axis('equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# Here we create a legend
# we'll plot empty lists with the desired size and label so we can include them in legend
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + ' km$^2$')
    
plt.legend(scatterpoints=1, frameon=True, labelspacing=1, title='City Area')
plt.title('California Cities: Area and Population');
```
