```python
import numpy as np
import pandas as pd

rng = np.random.RandomState(42)
df = pd.DataFrame(rng.rand(1000, 3), columns=['A', 'B', 'C'])
df.head()
```

```python
df.eval('D= A+B ',inplace=True)
df.head()
```

```python
df.query('D<1 and C<A')
```
